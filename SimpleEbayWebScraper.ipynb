```python
from urllib.request import urlopen
from bs4 import BeautifulSoup as bs
```

```python
base_url = "https://www.ebay.com/sch/i.html?_from=R40&_nkw="
request = "xbox"

minimum = 150
maximum = 250

url_separator = "&_sacat=0&_pgn="
minmax = "&_udlo="+str(minimum)+"&_udhi="+ str(maximum)
page_num = "1"

url = base_url + request + minmax + url_separator + page_num
```

```python
html = urlopen(url).read()
```

```python
print(html)
```

```python
soup = bs(html, "html.parser")
```

```python
print(soup)
```

```python
import re

for i in range(2, 11):
    print('Page ', page_num)
    for post in soup.find_all("li", {'class':re.compile('s-item.*')}):
        link = post.find_all("a", {"class": "s-item__link"})
        title = post.find_all("h3", {"class" : "s-item__title"})
        price = post.find_all("span", {"class":"s-item__price"})
        if title:
            print(title[0].text)
        if link:
            link = link[0]
            print(link['href'])
        if price:
            print(price[0].text+'\n')
        page_num = str(i)
        url = base_url + request + minmax + url_separator + page_num
        html = urlopen(url).read()
        soup = bs(html, "html.parser")
```
